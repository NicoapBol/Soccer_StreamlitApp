```python
events = sb.competition_events(
    country="International",
    division= "FIFA World Cup",
    season="2022",
    gender="male"
)

grouped_events = sb.competition_events(
    country="International",
    division= "FIFA World Cup",
    season="2022",
    split=True
)
```

```python
events = grouped_events["shots"]  # This is a list
events = pd.DataFrame(events)  # Convert to DataFrame
events = events[events["team"] == "Argentina"]
```

```python
matches=sb.matches(competition_id=43, season_id=106)
```

```python
matches = pd.DataFrame(matches)
matches = matches.drop(
    [        "referee", "home_managers", "away_managers", "data_version",
        "shot_fidelity_version", "xy_fidelity_version", "match_status_360",
        "last_updated", "last_updated_360", "match_week"
    ], 
    axis=1
)

matches.columns
```

```python
matches = matches[(matches["home_team"] == "Argentina") | (matches["away_team"] == "Argentina")]
matches["match_info"] = matches["home_team"] + " Vs. " + matches["away_team"] + " " + "(" + matches["match_date"] + ")"
matches["outcome"] = matches["home_team"] +" "+ matches["home_score"].astype(str) + " - " + matches["away_team"]+ " " + matches["away_score"].astype(str)
```

```python
matches.to_csv("world_cup22_matches.csv")
```

```python
events["penalty"] = events["shot_type"].apply(lambda x: "yes" if x == "penalty" else "no")
events[['X', 'Y']] = pd.DataFrame(events['location'].tolist(), index=events.index)
events["xG"] = events['shot_statsbomb_xg']
events["result"] = events['shot_outcome']
```

```python
events = events.merge(matches, on="match_id", how="left")
events.to_csv("arg_eventsWC22.csv", index=False, encoding="utf-8-sig")
```
